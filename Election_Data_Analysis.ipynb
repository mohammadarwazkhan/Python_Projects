## Q7. In how many constituencies, the difference in Total Votes between the Bharatiya Janata Party and the Samajwadi Party was less than 5000.

```python
# Filter for BJP and SP only
filtered = combined_election_data[combined_election_data['Party'].isin(['Bharatiya Janata Party', 'Samajwadi Party'])]
```

```python
filtered
```

```python
# Pivot table to get one row per constituency with BJP and SP votes as columns
pivot = filtered.pivot_table(index='ac_no', 
                             columns='Party', 
                             values='Total Votes', 
                             aggfunc='sum')
```

```python
pivot
```

```python
# Drop rows where one of the parties didn't contest (i.e., has NaN)
pivot = pivot.dropna(subset=['Bharatiya Janata Party', 'Samajwadi Party'])
```

```python
pivot
```

```python
# Calculate the vote difference
pivot['vote_diff'] = abs(pivot['Bharatiya Janata Party'] - pivot['Samajwadi Party'])
```

```python
pivot['vote_diff']
```

```python
# Find constituencies with vote difference less than 5000
close_contests = pivot[pivot['vote_diff'] < 5000]
```

```python
close_contests
```

```python
("Number of constituencies with vote difference < 5000:", len(close_contests))
```

## Q8. For the consituencies in part 7, in how many constituencies did the BJP lose (did not come first)

```python
# Get the list of those ac_no values
close_ac_nos = close_contests.index.tolist()
```

```python
close_ac_nos
```

```python
# From the original dataframe, get BJP candidates from those constituencies
bjp_in_close = combined_election_data[(combined_election_data['ac_no'].isin(close_ac_nos)) & (combined_election_data['Party'] == 'Bharatiya Janata Party')]
```

```python
bjp_in_close
```

```python
# Count how many of those BJP candidates did NOT have rank 1
bjp_losses = bjp_in_close[bjp_in_close['Rank'] != 1]
```

```python
print("BJP lost in", len(bjp_losses), "constituencies out of those with <5000 vote difference.")
```

## Q10. Calculate the district level vote % for each party. To calculate the vote % in a district. Add the total votes from all consituencies in that district for a party and divide it by the total votes in that district.

```python
# Total votes for each party in each district
party_district_votes = combined_election_data.groupby(['District', 'Party'])['Total Votes'].sum().reset_index()
```

```python
party_district_votes
```

```python
#  Total votes in each district (all parties)
district_total_votes = combined_election_data.groupby('District')['Total Votes'].sum().reset_index()
district_total_votes.rename(columns={'Total Votes': 'Total District Votes'}, inplace=True)
```

```python
district_total_votes
```

```python
#  Merge total district votes into party-level data
merged = party_district_votes.merge(district_total_votes, on='District')
```

```python
merged
```

```python
# Calculate vote percentage
merged['District Vote %'] = (merged['Total Votes'] / merged['Total District Votes']) * 100
merged['District Vote %'] = merged['District Vote %'].round(2)
```

```python
print(merged[['District', 'Party', 'District Vote %']])
```
